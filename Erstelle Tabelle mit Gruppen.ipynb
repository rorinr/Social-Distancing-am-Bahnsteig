```python
from ipynb.fs.full.whole_df import get_uppers, get_downers, get_whole_df, get_average_person_number
import os
import numpy as np
import itertools
import pandas as pd
import statistics
import math
```

```python
def getDistance(x1, x2, y1, y2):
    result = math.sqrt((x2 -x1)**2 + (y2 - y1)**2)
    return(result)
```

```python
keep_distance = []
average_number_of_persons = []
for filename in os.listdir(r"C:\Users\robin\Documents\Bachelorarbeit\Daten ZHDB"):    
    if "txt" in filename:
        raw_data = pd.read_csv("../Daten ZHDB/" + filename, skiprows = 4, sep = "\t", names = ["id", "frame", "x/m", "y/m", "h/m"])
        df = get_whole_df(filename, 5, 5)
        df.to_excel(str(filename[5:9]) + " Wartetrauben mit 5 Personen.xlsx")
```

```python
#Alle Abstnde berechnen die in Flche liegen
selectID = df
group_id_dataframe = pd.DataFrame(columns = ["GroupID", "doorPosition", "Einstiegsbeginn", "Einstiegsende", "Anzahl Personen Durchschnitt", "Abstand < 0.5", "0.5 < Abstand <= 1", "1 < Abstand <= 1.5", "1.5 < Abstand <= 2", "2 < Abstand <= 2.5", "2.5 < Abstand <= 3", "3 < Abstand"])
#Nur Personen innerhalb der Flche auswhlen
selectIDBorder = selectID.loc[selectID["inBorder"] == True]
for gruppe in selectIDBorder["GroupID"].unique():
    row_for_df = []
    borderDistance = []
    #Gruppe auswhlen
    selectGroup = selectIDBorder.loc[selectIDBorder["GroupID"] == gruppe, ["id", "frame_x", "x/m", "y/m_x", "doorPosition"]]
    begin = selectGroup["frame_x"].min()
    end = selectGroup["frame_x"].max()
    door_x = selectGroup["doorPosition"].mean()
    
    #Anzahl Personen auf Bahnsteig bekommmen
    df_number_person = raw_data.loc[raw_data["frame"].isin(selectGroup["frame_x"])]
    average_person_number = df_number_person["frame"].size / df_number_person["frame"].unique().size
    
    for frame in selectGroup["frame_x"].unique():
        selectFrame = selectGroup.loc[selectGroup["frame_x"] == frame]
        #Abstnde aller Kontakte berechnen
        for contact in itertools.combinations(selectFrame["id"], 2):
            firstPosition = selectFrame.loc[selectFrame["id"] == contact[0], ["x/m", "y/m_x"]]
            secondPosition = selectFrame.loc[selectFrame["id"] == contact[1], ["x/m", "y/m_x"]]
            distanz = getDistance(x1 = float(firstPosition["x/m"]), x2 = float(secondPosition["x/m"]), y1 = float(firstPosition["y/m_x"]), y2 = float(secondPosition["y/m_x"]))
            borderDistance.append(distanz)
    #Hufigkeiten der Kontakte berechnen
    halfMeter = [i for i in borderDistance if i <= 0.5]
    oneMeter = [i for i in borderDistance if (i <= 1) and (i > 0.5)]
    oneAndHalfMeter = [i for i in borderDistance if (i <= 1.5) and (i > 1)]
    twoMeter = [i for i in borderDistance if (i <= 2) and (i > 1.5)]
    twoAndHalfMeter = [i for i in borderDistance if (i <= 2.5) and (i > 2)]
    threeMeter = [i for i in borderDistance if (i <= 3) and (i > 2.5)]
    biggerThree = [i for i in borderDistance if i > 3]
    
    append_df = pd.DataFrame({"GroupID": [gruppe], "doorPosition": [door_x], "Einstiegsbeginn": [begin],
                              "Einstiegsende": [end], "Anzahl Personen Durchschnitt": [average_person_number],
                              "Abstand < 0.5": [len(halfMeter)], "0.5 < Abstand <= 1": [len(oneMeter)],
                              "1 < Abstand <= 1.5": [len(oneAndHalfMeter)], "1.5 < Abstand <= 2": [len(twoMeter)],
                              "2 < Abstand <= 2.5": [len(twoAndHalfMeter)], "2.5 < Abstand <= 3": [len(threeMeter)],
                              "3 < Abstand": [len(biggerThree)]})
    
    group_id_dataframe = group_id_dataframe.append(append_df)
group_id_dataframe
```
